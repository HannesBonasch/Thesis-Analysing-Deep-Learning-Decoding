```python
data_path = "F:/Masterthesis/Data/"
model_name = "deep"
model_folder = "ModelComparison"
task = "P3"
preprocessing = "heavy"

df = DataLoader.load_df(data_path, task, preprocessing)
data, labels = DataLoader.create_data_labels(df)
```

```python
# plot single epoch for all preprocessing levels

epoch = 0
subjectID = 1
pick = "FP1"
channels = ['FP1','F3','F7','FC3','C3','C5','P3','P7','P9','PO7',
            'PO3','O1','Oz','Pz','CPz','FP2','Fz','F4','F8','FC4',
            'FCz','Cz','C4','C6','P4','P8','P10','PO8','PO4','O2',]
pickID = channels.index(pick)
preprocessing = "light"
df = DataLoader.load_df(data_path, task, preprocessing)
e1 = df[(df["epoch"]==epoch) & (df["subjectID"]==subjectID)]["data"].item()[pickID]
preprocessing = "medium"
df = DataLoader.load_df(data_path, task, preprocessing)
e2 = df[(df["epoch"]==epoch) & (df["subjectID"]==subjectID)]["data"].item()[pickID]
preprocessing = "heavy"
df = DataLoader.load_df(data_path, task, preprocessing)
e3 = df[(df["epoch"]==epoch) & (df["subjectID"]==subjectID)]["data"].item()[pickID]

fig, ax1 = plt.subplots(figsize=(10, 3))
fig = sns.lineplot(data=[e1,e2,e3], dashes=None)
fig.set_xticks(np.arange(0,251,50))
fig.set_xticklabels([-0.2, 0, 0.2, 0.4, 0.6, 0.8])
fig.legend(['light', 'medium', 'heavy'])
fig.set(xlabel='Seconds', ylabel='µV');
```
