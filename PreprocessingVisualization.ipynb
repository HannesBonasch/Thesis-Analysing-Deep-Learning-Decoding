```python
epoch = 18
subjectID = 1
parameters["preprocessing"] = "light"
df = DataLoader.load_df(parameters)
e1 = df[(df["epoch"]==epoch) & (df["subjectID"]==subjectID)]["data"].item()[10]
parameters["preprocessing"] = "medium"
df = DataLoader.load_df(parameters)
e2 = df[(df["epoch"]==epoch) & (df["subjectID"]==subjectID)]["data"].item()[10]
parameters["preprocessing"] = "heavy"
df = DataLoader.load_df(parameters)
e3 = df[(df["epoch"]==epoch) & (df["subjectID"]==subjectID)]["data"].item()[10]
fig = sns.lineplot(data=[e1,e2,e3], dashes=None)
fig.set_xticks(np.arange(-0,251,50))
fig.set_xticklabels([-0.2, 0, 0.2, 0.4, 0.6, 0.8])
fig.legend(['light', 'medium', 'heavy'])
fig.set(xlabel='Seconds', ylabel='V');
```
